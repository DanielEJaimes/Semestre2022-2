1.Nombrar Columnas

```python
import pandas as pd
archivo = 'cancer_mama.csv'
datos = pd.read_csv(archivo)
columnas = []
for i in range(1,33,1):
  columnas.append("c"+str(i))
datos.columns=columnas
datos
```

2. Cambiar datos columna B

*   M = Maligno (Cambiar a 1)
*   B = Benigno (Cambiar a 0)

```python
datos['c2'] = datos['c2'].replace('M',1)
datos['c2'] = datos['c2'].replace('B',0)
datos
```

Matriz de correlación

```python
datos.corr()
```

3. Eliminar atípicos

```python
new_df = datos
caracteristicas = ['c3','c4','c5','c6']
for i in caracteristicas:
  x = i
  Q1 = new_df[x].quantile(0.25)
  Q3 = new_df[x].quantile(0.75)
  IQR = Q3 - Q1
  u_limit = Q3 + 0.5 * IQR
  l_limit = Q1 - 0.5 * IQR
  ubicacionNoAtipicos= (new_df[x] >= l_limit) & (new_df[x]<= u_limit)
  new_df = new_df[ubicacionNoAtipicos]

new_df
```

4. Entrenar modelo con las columnas seleccionadas acorde a la matriz de correlación

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

df_resultado = new_df[['c2']]
df_caracteristica = new_df[['c3','c4','c5','c6']]

x_train,x_test,y_train,y_test = train_test_split(df_caracteristica,df_resultado, test_size = 0.2) #test_size = porcentaje de datos para testear

clf = MLPClassifier(max_iter=200).fit(x_train,y_train)
clf.score(x_test,y_test)
```
